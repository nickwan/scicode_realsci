```python
import numpy as np
import pandas as pd

%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('Documents/hockey_data.xlsx', header=1)
df
```

```python
df.rename(columns={'Unnamed: 1': 'Team'}, inplace=1)
df
```

```python
df = df[df['Team'] != 'League Average']
```

```python
df['W'] = df['W'].astype(float)
df
```
